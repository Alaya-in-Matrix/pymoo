## C-TAEA


This algorithm is implemented based on <cite data-cite="ctaea"></cite> and targets constrained multi/many-objective problems. The algorithm is based on [Reference Directions](../misc/reference_directions.ipynb) which need to be provided when initializing the algorithm object.

It uses a two population approach to promote diversity.

```python
from pymoo.algorithms.ctaea import CTAEA
from pymoo.factory import get_problem, get_visualization, get_reference_directions
from pymoo.optimize import minimize

problem = get_problem("c3dtlz4")

algorithm = CTAEA(
    get_reference_directions("das-dennis", 3, n_partitions=12),
    seed=1
)

res = minimize(problem, algorithm, termination=('n_gen', 400))

get_visualization("scatter").add(res.F).show()
```
